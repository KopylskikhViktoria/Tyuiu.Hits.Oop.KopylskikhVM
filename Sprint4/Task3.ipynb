Создать базовый класс Product в C#, который представляет информацию о продуктах. На основе этого класса разработать 2–3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

<b>Требования к базовому классу Product:</b><br><br>

• <b>Атрибуты:</b> Название (Name), Цена (Price), Производитель (Manufacturer).<br><br>

• <b>Методы:</b><br>
    ○ GetInfo(): метод для получения информации о продукте в виде строки.<br>
    ○ Discount(): метод для применения скидки к цене продукта.<br>
    ○ Display(): метод для отображения информации о продукте на экране.<br><br>

<b>Требования к производным классам:</b><br><br>

1. Электроника (Electronics): Должен содержать дополнительные атрибуты, такие как Гарантийный срок (WarrantyPeriod). Метод Discount() должен быть переопределен для добавления логики учета гарантийного срока при применении скидки.<br><br>

2. Одежда (Clothing): Должен содержать дополнительные атрибуты, такие как Размер (Size). Метод Display() должен быть переопределен для добавления информации о размере при отображении информации о продукте.<br><br>

3. Книги (Books) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Автор (Author). Метод GetInfo() должен быть переопределен для включения информации об авторе в описание продукта.

```
public class Product
{
    // Закрытые поля класса
    private string name;
    private double price;
    private string manufacturer;
    private string category;
    private int stockQuantity;
    private double weight;
    private string article;
    private string description;

    // Свойство для названия продукта
    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    // Свойство для цены продукта
    public double Price
    {
        get { return price; }
        set
        {
            if (value >= 0)
                price = value;
        }
    }

    // Свойство для производителя
    public string Manufacturer
    {
        get { return manufacturer; }
        set { manufacturer = value; }
    }

    // Свойство для категории товара
    public string Category
    {
        get { return category; }
        set { category = value; }
    }

    // Свойство для количества товара на складе
    public int StockQuantity
    {
        get { return stockQuantity; }
        set
        {
            if (value >= 0)
                stockQuantity = value;
        }
    }

    // Свойство для веса товара
    public double Weight
    {
        get { return weight; }
        set
        {
            if (value >= 0)
                weight = value;
        }
    }

    // Новое свойство для артикула товара
    public string Article
    {
        get { return article; }
        set { article = value; }
    }

    // Новое свойство для описания товара
    public string Description
    {
        get { return description; }
        set { description = value; }
    }

    // Конструктор по умолчанию
    public Product()
    {
        Name = "Неизвестно";
        Price = 0;
        Manufacturer = "Неизвестно";
        Category = "Без категории";
        StockQuantity = 0;
        Weight = 0;
        Article = "Не указан";
        Description = "Нет описания";
    }

    // Конструктор с параметрами
    public Product(string name, double price, string manufacturer, string category, int stockQuantity, double weight, string article, string description)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
        Category = category;
        StockQuantity = stockQuantity;
        Weight = weight;
        Article = article;
        Description = description;
    }

    // Возвращает основную информацию о продукте
    public virtual string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price}, Производитель: {Manufacturer}, Категория: {Category}, Количество: {StockQuantity}, Вес: {Weight}, Артикул: {Article}, Описание: {Description}";
    }

    // Применяет скидку к цене продукта
    public virtual void Discount(double percent)
    {
        Price = Price - (Price * percent / 100);
    }

    // Перегруженный метод скидки с выводом сообщения
    public void Discount(double percent, bool showMessage)
    {
        Discount(percent);

        if (showMessage)
        {
            Console.WriteLine($"К товару {Name} применена скидка {percent}%.");
        }
    }

    // Выводит информацию о продукте на экран
    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }

    // Перегруженный метод вывода с заголовком
    public void Display(string title)
    {
        Console.WriteLine(title);
        Display();
    }

    // Сравнивает цену текущего объекта с другим объектом Product
    public void ComparePrice(Product other)
    {
        if (Price > other.Price)
        {
            Console.WriteLine($"{Name} дороже, чем {other.Name}");
        }
        else if (Price < other.Price)
        {
            Console.WriteLine($"{Name} дешевле, чем {other.Name}");
        }
        else
        {
            Console.WriteLine($"{Name} и {other.Name} стоят одинаково");
        }
    }

    // Проверяет, есть ли товар в наличии
    public bool IsAvailable()
    {
        return StockQuantity > 0;
    }

    // Пополняет запас товара на складе
    public void Restock(int amount)
    {
        if (amount > 0)
        {
            StockQuantity += amount;
        }
    }

    // Изменяет цену товара
    public void ChangePrice(double newPrice)
    {
        if (newPrice >= 0)
        {
            Price = newPrice;
        }
    }

    // Уменьшает количество товара на складе
    public void Sell(int amount)
    {
        if (amount > 0 && amount <= StockQuantity)
        {
            StockQuantity -= amount;
        }
    }

    // Новый метод для изменения описания товара
    public void UpdateDescription(string newDescription)
    {
        Description = newDescription;
    }
}

public class Electronics : Product, IWarrantyService
{
    // Закрытые поля для характеристик электроники
    private int warrantyPeriod;
    private int power;
    private bool isWireless;
    private string energyClass;

    // Свойство для гарантийного срока
    public int WarrantyPeriod
    {
        get { return warrantyPeriod; }
        set
        {
            if (value >= 0)
                warrantyPeriod = value;
        }
    }

    // Свойство для мощности
    public int Power
    {
        get { return power; }
        set
        {
            if (value >= 0)
                power = value;
        }
    }

    // Свойство для признака беспроводного устройства
    public bool IsWireless
    {
        get { return isWireless; }
        set { isWireless = value; }
    }

    // Новое свойство для класса энергопотребления
    public string EnergyClass
    {
        get { return energyClass; }
        set { energyClass = value; }
    }

    // Конструктор по умолчанию
    public Electronics() : base()
    {
        WarrantyPeriod = 0;
        Power = 0;
        IsWireless = false;
        EnergyClass = "Не указан";
    }

    // Конструктор с параметрами
    public Electronics(string name, double price, string manufacturer, string category, int stockQuantity, double weight,
        string article, string description, int warrantyPeriod, int power, bool isWireless, string energyClass)
        : base(name, price, manufacturer, category, stockQuantity, weight, article, description)
    {
        WarrantyPeriod = warrantyPeriod;
        Power = power;
        IsWireless = isWireless;
        EnergyClass = energyClass;
    }

    // Переопределение метода скидки с учетом гарантийного срока
    public override void Discount(double percent)
    {
        if (WarrantyPeriod > 24)
        {
            percent = percent - 2;
        }

        if (percent < 0)
        {
            percent = 0;
        }

        Price = Price - (Price * percent / 100);
    }

    // Переопределение вывода информации
    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Гарантия: {WarrantyPeriod}, Мощность: {Power}, Беспроводной: {IsWireless}, Энергокласс: {EnergyClass}");
    }

    // Включает электронное устройство
    public void TurnOn()
    {
        Console.WriteLine($"{Name} включён.");
    }

    // Новый метод для проверки мощности
    public void CheckPower()
    {
        Console.WriteLine($"Мощность устройства {Name}: {Power} Вт.");
    }

    // Явная реализация интерфейса гарантии
    void IWarrantyService.RegisterWarranty()
    {
        Console.WriteLine($"Гарантия на товар {Name} зарегистрирована на {WarrantyPeriod} мес.");
    }
}

public class Clothing : Product, IFittingService
{
    // Закрытые поля для одежды
    private string size;
    private string material;
    private string season;
    private string style;

    // Свойство для размера
    public string Size
    {
        get { return size; }
        set { size = value; }
    }

    // Свойство для материала
    public string Material
    {
        get { return material; }
        set { material = value; }
    }

    // Свойство для сезона
    public string Season
    {
        get { return season; }
        set { season = value; }
    }

    // Новое свойство для стиля одежды
    public string Style
    {
        get { return style; }
        set { style = value; }
    }

    // Конструктор по умолчанию
    public Clothing() : base()
    {
        Size = "Не указан";
        Material = "Не указан";
        Season = "Не указан";
        Style = "Не указан";
    }

    // Конструктор с параметрами
    public Clothing(string name, double price, string manufacturer, string category, int stockQuantity, double weight,
        string article, string description, string size, string material, string season, string style)
        : base(name, price, manufacturer, category, stockQuantity, weight, article, description)
    {
        Size = size;
        Material = material;
        Season = season;
        Style = style;
    }

    // Переопределение вывода с добавлением размера
    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Размер: {Size}, Материал: {Material}, Сезон: {Season}, Стиль: {Style}");
    }

    // Примерка одежды
    public void TryOn()
    {
        Console.WriteLine($"{Name} можно примерить. Размер: {Size}");
    }

    // Новый метод для показа стиля одежды
    public void ShowStyle()
    {
        Console.WriteLine($"Стиль товара {Name}: {Style}");
    }

    // Явная реализация интерфейса примерки
    void IFittingService.StartFitting()
    {
        Console.WriteLine($"Примерка товара {Name} началась.");
    }
}

public class Books : Product, IReadingService
{
    // Закрытые поля для книги
    private string author;
    private int pages;
    private string genre;
    private string language;

    // Свойство для автора
    public string Author
    {
        get { return author; }
        set { author = value; }
    }

    // Свойство для количества страниц
    public int Pages
    {
        get { return pages; }
        set
        {
            if (value >= 0)
                pages = value;
        }
    }

    // Свойство для жанра
    public string Genre
    {
        get { return genre; }
        set { genre = value; }
    }

    // Новое свойство для языка книги
    public string Language
    {
        get { return language; }
        set { language = value; }
    }

    // Конструктор по умолчанию
    public Books() : base()
    {
        Author = "Неизвестен";
        Pages = 0;
        Genre = "Не указан";
        Language = "Не указан";
    }

    // Конструктор с параметрами
    public Books(string name, double price, string manufacturer, string category, int stockQuantity, double weight,
        string article, string description, string author, int pages, string genre, string language)
        : base(name, price, manufacturer, category, stockQuantity, weight, article, description)
    {
        Author = author;
        Pages = pages;
        Genre = genre;
        Language = language;
    }

    // Переопределение метода для добавления автора в описание
    public override string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price}, Производитель: {Manufacturer}, Категория: {Category}, Количество: {StockQuantity}, Вес: {Weight}, Артикул: {Article}, Описание: {Description}, Автор: {Author}, Страниц: {Pages}, Жанр: {Genre}, Язык: {Language}";
    }

    // Открывает книгу для чтения
    public void Read()
    {
        Console.WriteLine($"Вы открыли книгу \"{Name}\".");
    }

    // Новый метод для показа количества страниц
    public void ShowPages()
    {
        Console.WriteLine($"В книге {Name} {Pages} страниц.");
    }

    // Явная реализация интерфейса чтения
    void IReadingService.OpenPreview()
    {
        Console.WriteLine($"Открыт фрагмент книги \"{Name}\".");
    }
}

// Интерфейс для явной реализации в классе Electronics
public interface IWarrantyService
{
    void RegisterWarranty();
}

// Интерфейс для явной реализации в классе Clothing
public interface IFittingService
{
    void StartFitting();
}

// Интерфейс для явной реализации в классе Books
public interface IReadingService
{
    void OpenPreview();
}

// Интерфейс для управления товарами
public interface IProductManager
{
    void ProcessDiscount(Product product, double percent);
    void ShowProduct(Product product);
}

// Класс для управления зависимостями через интерфейс
public class ProductManager : IProductManager
{
    public void ProcessDiscount(Product product, double percent)
    {
        product.Discount(percent);
    }

    public void ShowProduct(Product product)
    {
        product.Display();
    }
}

// Класс сервиса получает зависимость через конструктор
public class StoreService
{
    private IProductManager productManager;

    public StoreService(IProductManager productManager)
    {
        this.productManager = productManager;
    }

    // Метод использует внедрённую зависимость
    public void ServeProduct(Product product, double percent)
    {
        productManager.ShowProduct(product);
        productManager.ProcessDiscount(product, percent);
        Console.WriteLine("После обработки товара:");
        productManager.ShowProduct(product);
    }
}
```

```
Electronics laptop = new Electronics("Ноутбук", 80000, "Lenovo", "Электроника", 5, 2.4, "E-101", "Игровой ноутбук", 36, 120, false, "A");
Clothing tshirt = new Clothing("Футболка", 1500, "Nike", "Одежда", 20, 0.3, "C-205", "Летняя футболка", "M", "Хлопок", "Лето", "Casual");
Books book = new Books("Преступление и наказание", 700, "Эксмо", "Книги", 12, 0.5, "B-777", "Классический роман", "Ф. М. Достоевский", 672, "Роман", "Русский");

// Управление зависимостями через внедрение объекта менеджера
IProductManager manager = new ProductManager();
StoreService storeService = new StoreService(manager);

Console.WriteLine("Работа с электроникой:");
laptop.TurnOn();
laptop.CheckPower();
storeService.ServeProduct(laptop, 10);

Console.WriteLine();

// Использование явной реализации интерфейса
IWarrantyService warrantyService = laptop;
warrantyService.RegisterWarranty();

Console.WriteLine();

Console.WriteLine("Работа с одеждой:");
tshirt.TryOn();
tshirt.ShowStyle();
storeService.ServeProduct(tshirt, 5);

Console.WriteLine();

IFittingService fittingService = tshirt;
fittingService.StartFitting();

Console.WriteLine();

Console.WriteLine("Работа с книгой:");
book.Read();
book.ShowPages();
storeService.ServeProduct(book, 3);

Console.WriteLine();

IReadingService readingService = book;
readingService.OpenPreview();

Console.WriteLine();

Console.WriteLine("Сравнение товаров:");
laptop.ComparePrice(tshirt);
book.ComparePrice(laptop);

Console.WriteLine();

Console.WriteLine("Проверка наличия и продаж:");
Console.WriteLine($"Футболка в наличии: {tshirt.IsAvailable()}");
tshirt.Sell(2);
Console.WriteLine($"Количество футболок после продажи: {tshirt.StockQuantity}");
book.Restock(5);
Console.WriteLine($"Количество книг после пополнения: {book.StockQuantity}");
```
