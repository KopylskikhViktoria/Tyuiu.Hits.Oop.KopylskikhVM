<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Product в C#, который представляет информацию о продуктах. На основе этого класса разработать 2–3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

<b>Требования к базовому классу Product:</b><br><br>

• <b>Атрибуты:</b> Название (Name), Цена (Price), Производитель (Manufacturer).<br><br>

• <b>Методы:</b><br>
    ○ GetInfo(): метод для получения информации о продукте в виде строки.<br>
    ○ Discount(): метод для применения скидки к цене продукта.<br>
    ○ Display(): метод для отображения информации о продукте на экране.<br><br>

<b>Требования к производным классам:</b><br><br>

1. Электроника (Electronics): Должен содержать дополнительные атрибуты, такие как Гарантийный срок (WarrantyPeriod). Метод Discount() должен быть переопределен для добавления логики учета гарантийного срока при применении скидки.<br><br>

2. Одежда (Clothing): Должен содержать дополнительные атрибуты, такие как Размер (Size). Метод Display() должен быть переопределен для добавления информации о размере при отображении информации о продукте.<br><br>

3. Книги (Books) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Автор (Author). Метод GetInfo() должен быть переопределен для включения информации об авторе в описание продукта.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) исользуйтие в проекте коллекции, делегаты, события.

```
public delegate void ProductDelegate(Product product);

public class Product
{
    public string Name { get; set; }
    public double Price { get; set; }
    public string Manufacturer { get; set; }

    // Новые атрибуты базового класса
    public string Category { get; set; }
    public int Quantity { get; set; }
    public double Rating { get; set; }

    // Событие для уведомления об изменении цены
    public delegate void ProductHandler(string message);
    public event ProductHandler ProductUpdated;

    public Product(string name, double price, string manufacturer, string category, int quantity, double rating)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
        Category = category;
        Quantity = quantity;
        Rating = rating;
    }

    // Защищённый метод для вызова события
    protected void OnProductUpdated(string message)
    {
        ProductUpdated?.Invoke(message);
    }

    // Возвращает основную информацию о продукте
    public virtual string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price}, Производитель: {Manufacturer}, Категория: {Category}, Количество: {Quantity}, Рейтинг: {Rating}";
    }

    // Применяет скидку к цене продукта
    public virtual void Discount(double percent)
    {
        Price = Price - (Price * percent / 100);
        OnProductUpdated($"Цена товара {Name} изменена. Новая цена: {Price}");
    }

    // Выводит информацию о продукте на экран
    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }

    // Новый метод для пополнения количества товара
    public void Restock(int amount)
    {
        Quantity += amount;
    }

    // Новый метод для продажи товара
    public void Sell(int amount)
    {
        if (amount <= Quantity)
        {
            Quantity -= amount;
        }
    }

    // Новый метод для изменения рейтинга
    public void UpdateRating(double newRating)
    {
        Rating = newRating;
    }

    // Новый метод для изменения цены
    public void UpdatePrice(double newPrice)
    {
        Price = newPrice;
        OnProductUpdated($"Цена товара {Name} обновлена вручную. Новая цена: {Price}");
    }
}

public class Electronics : Product
{
    public int WarrantyPeriod { get; set; }

    // Новый атрибут класса Electronics
    public int Power { get; set; }

    public Electronics(string name, double price, string manufacturer, string category, int quantity, double rating, int warrantyPeriod, int power)
        : base(name, price, manufacturer, category, quantity, rating)
    {
        WarrantyPeriod = warrantyPeriod;
        Power = power;
    }

    // Переопределение метода скидки с учетом гарантийного срока
    public override void Discount(double percent)
    {
        if (WarrantyPeriod > 24)
        {
            percent = percent - 2;
        }

        if (percent < 0)
        {
            percent = 0;
        }

        Price = Price - (Price * percent / 100);
        OnProductUpdated($"Скидка для электроники {Name} применена. Новая цена: {Price}");
    }

    // Новый метод для включения устройства
    public void TurnOn()
    {
        Console.WriteLine($"{Name} включён.");
    }
}

public class Clothing : Product
{
    public string Size { get; set; }

    // Новый атрибут класса Clothing
    public string Material { get; set; }

    public Clothing(string name, double price, string manufacturer, string category, int quantity, double rating, string size, string material)
        : base(name, price, manufacturer, category, quantity, rating)
    {
        Size = size;
        Material = material;
    }

    // Переопределение вывода с добавлением размера
    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Размер: {Size}, Материал: {Material}");
    }

    // Новый метод для примерки
    public void TryOn()
    {
        Console.WriteLine($"{Name} можно примерить.");
    }
}

public class Books : Product
{
    public string Author { get; set; }

    // Новый атрибут класса Books
    public int Pages { get; set; }

    public Books(string name, double price, string manufacturer, string category, int quantity, double rating, string author, int pages)
        : base(name, price, manufacturer, category, quantity, rating)
    {
        Author = author;
        Pages = pages;
    }

    // Переопределение метода для добавления автора в описание
    public override string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price}, Производитель: {Manufacturer}, Категория: {Category}, Количество: {Quantity}, Рейтинг: {Rating}, Автор: {Author}, Страниц: {Pages}";
    }

    // Новый метод для чтения книги
    public void ReadSample()
    {
        Console.WriteLine($"Открыт фрагмент книги {Name}.");
    }
}
```

```
Electronics laptop = new Electronics("Ноутбук", 80000, "Lenovo", "Электроника", 5, 4.8, 36, 120);
Clothing tshirt = new Clothing("Футболка", 1500, "Nike", "Одежда", 20, 4.5, "M", "Хлопок");
Books book = new Books("Преступление и наказание", 700, "Эксмо", "Книги", 12, 4.9, "Ф. М. Достоевский", 672);

// Подписка на события
laptop.ProductUpdated += OnProductUpdated;
tshirt.ProductUpdated += OnProductUpdated;
book.ProductUpdated += OnProductUpdated;

Console.WriteLine("Информация об электронике:");
laptop.Display();
laptop.TurnOn();
laptop.Discount(10);
Console.WriteLine("После скидки:");
laptop.Display();

Console.WriteLine("\nИнформация об одежде:");
tshirt.Display();
tshirt.TryOn();
tshirt.Discount(15);
Console.WriteLine("После скидки:");
tshirt.Display();

Console.WriteLine("\nИнформация о книге:");
book.Display();
book.ReadSample();
book.Discount(5);
Console.WriteLine("После скидки:");
book.Display();

Console.WriteLine("\n=== Коллекция List<Product> ===");

// Коллекция товаров
List<Product> products = new List<Product>()
{
    laptop,
    tshirt,
    book
};

foreach (Product product in products)
{
    product.Display();
    Console.WriteLine();
}

Console.WriteLine("=== Коллекция Dictionary<string, Product> ===");

// Словарь товаров
Dictionary<string, Product> productDictionary = new Dictionary<string, Product>();
productDictionary.Add("E001", laptop);
productDictionary.Add("C001", tshirt);
productDictionary.Add("B001", book);

Console.WriteLine("Товар по ключу B001:");
productDictionary["B001"].Display();

Console.WriteLine("\n=== Делегаты ===");

// Делегаты для вывода информации
ProductDelegate showInfo = p => p.Display();

showInfo(laptop);
Console.WriteLine();
showInfo(tshirt);
Console.WriteLine();
showInfo(book);

Console.WriteLine("\n=== Работа с новыми методами ===");

laptop.Restock(3);
Console.WriteLine($"Количество ноутбуков после пополнения: {laptop.Quantity}");

tshirt.Sell(2);
Console.WriteLine($"Количество футболок после продажи: {tshirt.Quantity}");

book.UpdateRating(5.0);
Console.WriteLine($"Новый рейтинг книги: {book.Rating}");

book.UpdatePrice(650);

// Обработчик события
static void OnProductUpdated(string message)
{
    Console.WriteLine($"[Событие] {message}");
}
```
