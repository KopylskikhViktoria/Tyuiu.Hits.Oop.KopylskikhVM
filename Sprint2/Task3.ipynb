<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Product в C#, который представляет информацию о продуктах. На основе этого класса разработать 2–3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

<b>Требования к базовому классу Product:</b><br><br>

• <b>Атрибуты:</b> Название (Name), Цена (Price), Производитель (Manufacturer).<br><br>

• <b>Методы:</b><br>
    ○ GetInfo(): метод для получения информации о продукте в виде строки.<br>
    ○ Discount(): метод для применения скидки к цене продукта.<br>
    ○ Display(): метод для отображения информации о продукте на экране.<br><br>

<b>Требования к производным классам:</b><br><br>

1. Электроника (Electronics): Должен содержать дополнительные атрибуты, такие как Гарантийный срок (WarrantyPeriod). Метод Discount() должен быть переопределен для добавления логики учета гарантийного срока при применении скидки.<br><br>

2. Одежда (Clothing): Должен содержать дополнительные атрибуты, такие как Размер (Size). Метод Display() должен быть переопределен для добавления информации о размере при отображении информации о продукте.<br><br>

3. Книги (Books) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Автор (Author). Метод GetInfo() должен быть переопределен для включения информации об авторе в описание продукта.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

<h2 style="color:DodgerBlue">Реализация:</h2>

```
public class Product
{
    public string Name { get; set; }
    public double Price { get; set; }
    public string Manufacturer { get; set; }

    // Новые атрибуты базового класса
    public string Category { get; set; }
    public int Quantity { get; set; }
    public double Weight { get; set; }

    public Product(string name, double price, string manufacturer)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
        Category = "Не указана";
        Quantity = 0;
        Weight = 0;
    }

    public Product(string name, double price, string manufacturer, string category, int quantity, double weight)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
        Category = category;
        Quantity = quantity;
        Weight = weight;
    }

    // Возвращает основную информацию о продукте
    public virtual string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price}, Производитель: {Manufacturer}, Категория: {Category}, Количество: {Quantity}, Вес: {Weight}";
    }

    // Применяет скидку к цене продукта
    public virtual void Discount(double percent)
    {
        Price = Price - (Price * percent / 100);
    }

    // Выводит информацию о продукте на экран
    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }

    // Новый метод для пополнения товара
    public void Restock(int amount)
    {
        Quantity += amount;
    }

    // Новый метод для продажи товара
    public void Sell(int amount)
    {
        if (amount <= Quantity)
        {
            Quantity -= amount;
        }
    }

    // Новый метод для изменения веса
    public void ChangeWeight(double newWeight)
    {
        Weight = newWeight;
    }
}

public class Electronics : Product
{
    public int WarrantyPeriod { get; set; }

    // Новый атрибут класса Electronics
    public int Power { get; set; }

    public Electronics(string name, double price, string manufacturer, int warrantyPeriod)
        : base(name, price, manufacturer)
    {
        WarrantyPeriod = warrantyPeriod;
        Power = 0;
    }

    public Electronics(string name, double price, string manufacturer, string category, int quantity, double weight, int warrantyPeriod, int power)
        : base(name, price, manufacturer, category, quantity, weight)
    {
        WarrantyPeriod = warrantyPeriod;
        Power = power;
    }

    // Переопределение метода скидки с учетом гарантийного срока
    public override void Discount(double percent)
    {
        if (WarrantyPeriod > 24)
        {
            percent = percent - 2;
        }

        if (percent < 0)
        {
            percent = 0;
        }

        Price = Price - (Price * percent / 100);
    }

    // Новый метод для включения устройства
    public void TurnOn()
    {
        Console.WriteLine($"{Name} включён.");
    }
}

public class Clothing : Product
{
    public string Size { get; set; }

    // Новый атрибут класса Clothing
    public string Material { get; set; }

    public Clothing(string name, double price, string manufacturer, string size)
        : base(name, price, manufacturer)
    {
        Size = size;
        Material = "Не указан";
    }

    public Clothing(string name, double price, string manufacturer, string category, int quantity, double weight, string size, string material)
        : base(name, price, manufacturer, category, quantity, weight)
    {
        Size = size;
        Material = material;
    }

    // Переопределение вывода с добавлением размера
    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Размер: {Size}, Материал: {Material}");
    }

    // Новый метод для примерки
    public void TryOn()
    {
        Console.WriteLine($"{Name} можно примерить.");
    }
}

public class Books : Product
{
    public string Author { get; set; }

    // Новый атрибут класса Books
    public int Pages { get; set; }

    public Books(string name, double price, string manufacturer, string author)
        : base(name, price, manufacturer)
    {
        Author = author;
        Pages = 0;
    }

    public Books(string name, double price, string manufacturer, string category, int quantity, double weight, string author, int pages)
        : base(name, price, manufacturer, category, quantity, weight)
    {
        Author = author;
        Pages = pages;
    }

    // Переопределение метода для добавления автора в описание
    public override string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price}, Производитель: {Manufacturer}, Категория: {Category}, Количество: {Quantity}, Вес: {Weight}, Автор: {Author}, Страниц: {Pages}";
    }

    // Новый метод для чтения книги
    public void ReadSample()
    {
        Console.WriteLine($"Открыт фрагмент книги {Name}.");
    }
}

// Интерфейсы для множественного наследования
public interface IChargeable
{
    void Charge();
}

public interface IConnectable
{
    void Connect();
}

// Сложное наследование: SmartPhone наследуется от Electronics
// Множественное наследование: через интерфейсы IChargeable и IConnectable
public class SmartPhone : Electronics, IChargeable, IConnectable
{
    public string OperatingSystem { get; set; }
    public int Memory { get; set; }

    public SmartPhone(string name, double price, string manufacturer, string category, int quantity, double weight,
        int warrantyPeriod, int power, string operatingSystem, int memory)
        : base(name, price, manufacturer, category, quantity, weight, warrantyPeriod, power)
    {
        OperatingSystem = operatingSystem;
        Memory = memory;
    }

    // Переопределение вывода информации о смартфоне
    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Гарантия: {WarrantyPeriod}, Мощность: {Power}, ОС: {OperatingSystem}, Память: {Memory}");
    }

    public void Charge()
    {
        Console.WriteLine($"{Name} заряжается.");
    }

    public void Connect()
    {
        Console.WriteLine($"{Name} подключён к сети.");
    }
}
```

```
// Простое наследование
Electronics laptop = new Electronics("Ноутбук", 80000, "Lenovo", "Электроника", 5, 2.4, 36, 120);
Clothing tshirt = new Clothing("Футболка", 1500, "Nike", "Одежда", 20, 0.3, "M", "Хлопок");
Books book = new Books("Преступление и наказание", 700, "Эксмо", "Книги", 12, 0.5, "Ф. М. Достоевский", 672);

// Сложное и множественное наследование
SmartPhone phone = new SmartPhone("Смартфон", 65000, "Samsung", "Электроника", 8, 0.2, 24, 25, "Android", 256);

Console.WriteLine("Информация об электронике:");
laptop.Display();
laptop.TurnOn();
laptop.Discount(10);
Console.WriteLine("После скидки:");
laptop.Display();

Console.WriteLine("\nИнформация об одежде:");
tshirt.Display();
tshirt.TryOn();
tshirt.Sell(2);
Console.WriteLine($"Количество после продажи: {tshirt.Quantity}");

Console.WriteLine("\nИнформация о книге:");
book.Display();
book.ReadSample();
book.Restock(5);
Console.WriteLine($"Количество после пополнения: {book.Quantity}");

Console.WriteLine("\nИнформация о смартфоне:");
phone.Display();
phone.TurnOn();
phone.Charge();
phone.Connect();
phone.Discount(7);
Console.WriteLine("После скидки:");
phone.Display();
```
