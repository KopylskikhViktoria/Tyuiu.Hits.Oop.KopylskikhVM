<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Product в C#, который представляет информацию о продуктах. На основе этого класса разработать 2–3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

<b>Требования к базовому классу Product:</b><br><br>

• <b>Атрибуты:</b> Название (Name), Цена (Price), Производитель (Manufacturer).<br><br>

• <b>Методы:</b><br>
    ○ GetInfo(): метод для получения информации о продукте в виде строки.<br>
    ○ Discount(): метод для применения скидки к цене продукта.<br>
    ○ Display(): метод для отображения информации о продукте на экране.<br><br>

<b>Требования к производным классам:</b><br><br>

1. Электроника (Electronics): Должен содержать дополнительные атрибуты, такие как Гарантийный срок (WarrantyPeriod). Метод Discount() должен быть переопределен для добавления логики учета гарантийного срока при применении скидки.<br><br>

2. Одежда (Clothing): Должен содержать дополнительные атрибуты, такие как Размер (Size). Метод Display() должен быть переопределен для добавления информации о размере при отображении информации о продукте.<br><br>

3. Книги (Books) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Автор (Author). Метод GetInfo() должен быть переопределен для включения информации об авторе в описание продукта.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
public class Product
{
    // Закрытые поля класса
    private string name;
    private double price;
    private string manufacturer;
    private string category;
    private int stockQuantity;
    private double weight;

    // Свойство для названия продукта
    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    // Свойство для цены продукта
    public double Price
    {
        get { return price; }
        set
        {
            if (value >= 0)
                price = value;
        }
    }

    // Свойство для производителя
    public string Manufacturer
    {
        get { return manufacturer; }
        set { manufacturer = value; }
    }

    // Свойство для категории товара
    public string Category
    {
        get { return category; }
        set { category = value; }
    }

    // Свойство для количества товара на складе
    public int StockQuantity
    {
        get { return stockQuantity; }
        set
        {
            if (value >= 0)
                stockQuantity = value;
        }
    }

    // Свойство для веса товара
    public double Weight
    {
        get { return weight; }
        set
        {
            if (value >= 0)
                weight = value;
        }
    }

    // Конструктор по умолчанию
    public Product()
    {
        Name = "Неизвестно";
        Price = 0;
        Manufacturer = "Неизвестно";
        Category = "Без категории";
        StockQuantity = 0;
        Weight = 0;
    }

    // Конструктор с параметрами
    public Product(string name, double price, string manufacturer, string category, int stockQuantity, double weight)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
        Category = category;
        StockQuantity = stockQuantity;
        Weight = weight;
    }

    // Возвращает основную информацию о продукте
    public virtual string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price}, Производитель: {Manufacturer}, Категория: {Category}, Количество: {StockQuantity}, Вес: {Weight}";
    }

    // Применяет скидку к цене продукта
    public virtual void Discount(double percent)
    {
        Price = Price - (Price * percent / 100);
    }

    // Выводит информацию о продукте на экран
    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }

    // Сравнивает цену текущего объекта с другим объектом Product
    public void ComparePrice(Product other)
    {
        if (Price > other.Price)
        {
            Console.WriteLine($"{Name} дороже, чем {other.Name}");
        }
        else if (Price < other.Price)
        {
            Console.WriteLine($"{Name} дешевле, чем {other.Name}");
        }
        else
        {
            Console.WriteLine($"{Name} и {other.Name} стоят одинаково");
        }
    }

    // Проверяет, есть ли товар в наличии
    public bool IsAvailable()
    {
        return StockQuantity > 0;
    }

    // Пополняет запас товара на складе
    public void Restock(int amount)
    {
        if (amount > 0)
        {
            StockQuantity += amount;
        }
    }

    // Изменяет цену товара
    public void ChangePrice(double newPrice)
    {
        if (newPrice >= 0)
        {
            Price = newPrice;
        }
    }
}

public class Electronics : Product
{
    // Закрытые поля для характеристик электроники
    private int warrantyPeriod;
    private int power;
    private bool isWireless;

    // Свойство для гарантийного срока
    public int WarrantyPeriod
    {
        get { return warrantyPeriod; }
        set
        {
            if (value >= 0)
                warrantyPeriod = value;
        }
    }

    // Свойство для мощности
    public int Power
    {
        get { return power; }
        set
        {
            if (value >= 0)
                power = value;
        }
    }

    // Свойство для признака беспроводного устройства
    public bool IsWireless
    {
        get { return isWireless; }
        set { isWireless = value; }
    }

    // Конструктор по умолчанию
    public Electronics() : base()
    {
        WarrantyPeriod = 0;
        Power = 0;
        IsWireless = false;
    }

    // Конструктор с параметрами
    public Electronics(string name, double price, string manufacturer, string category, int stockQuantity, double weight,
        int warrantyPeriod, int power, bool isWireless)
        : base(name, price, manufacturer, category, stockQuantity, weight)
    {
        WarrantyPeriod = warrantyPeriod;
        Power = power;
        IsWireless = isWireless;
    }

    // Переопределение метода скидки с учетом гарантийного срока
    public override void Discount(double percent)
    {
        if (WarrantyPeriod > 24)
        {
            percent = percent - 2;
        }

        if (percent < 0)
        {
            percent = 0;
        }

        Price = Price - (Price * percent / 100);
    }

    // Включает электронное устройство
    public void TurnOn()
    {
        Console.WriteLine($"{Name} включён.");
    }
}

public class Clothing : Product
{
    // Закрытые поля для одежды
    private string size;
    private string material;
    private string season;

    // Свойство для размера
    public string Size
    {
        get { return size; }
        set { size = value; }
    }

    // Свойство для материала
    public string Material
    {
        get { return material; }
        set { material = value; }
    }

    // Свойство для сезона
    public string Season
    {
        get { return season; }
        set { season = value; }
    }

    // Конструктор по умолчанию
    public Clothing() : base()
    {
        Size = "Не указан";
        Material = "Не указан";
        Season = "Не указан";
    }

    // Конструктор с параметрами
    public Clothing(string name, double price, string manufacturer, string category, int stockQuantity, double weight,
        string size, string material, string season)
        : base(name, price, manufacturer, category, stockQuantity, weight)
    {
        Size = size;
        Material = material;
        Season = season;
    }

    // Переопределение вывода с добавлением размера
    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Размер: {Size}, Материал: {Material}, Сезон: {Season}");
    }

    // Примерка одежды
    public void TryOn()
    {
        Console.WriteLine($"{Name} можно примерить. Размер: {Size}");
    }
}

public class Books : Product
{
    // Закрытые поля для книги
    private string author;
    private int pages;
    private string genre;

    // Свойство для автора
    public string Author
    {
        get { return author; }
        set { author = value; }
    }

    // Свойство для количества страниц
    public int Pages
    {
        get { return pages; }
        set
        {
            if (value >= 0)
                pages = value;
        }
    }

    // Свойство для жанра
    public string Genre
    {
        get { return genre; }
        set { genre = value; }
    }

    // Конструктор по умолчанию
    public Books() : base()
    {
        Author = "Неизвестен";
        Pages = 0;
        Genre = "Не указан";
    }

    // Конструктор с параметрами
    public Books(string name, double price, string manufacturer, string category, int stockQuantity, double weight,
        string author, int pages, string genre)
        : base(name, price, manufacturer, category, stockQuantity, weight)
    {
        Author = author;
        Pages = pages;
        Genre = genre;
    }

    // Переопределение метода для добавления автора в описание
    public override string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price}, Производитель: {Manufacturer}, Категория: {Category}, Количество: {StockQuantity}, Вес: {Weight}, Автор: {Author}, Страниц: {Pages}, Жанр: {Genre}";
    }

    // Открывает книгу для чтения
    public void Read()
    {
        Console.WriteLine($"Вы открыли книгу \"{Name}\".");
    }
}

// Интерфейсы для множественного наследования
public interface IChargeable
{
    void Charge();
}

public interface IConnectable
{
    void ConnectToInternet();
}

// Сложное и множественное наследование
public class SmartPhone : Electronics, IChargeable, IConnectable
{
    // Дополнительные поля смартфона
    private string operatingSystem;
    private int memory;

    // Свойство для операционной системы
    public string OperatingSystem
    {
        get { return operatingSystem; }
        set { operatingSystem = value; }
    }

    // Свойство для объема памяти
    public int Memory
    {
        get { return memory; }
        set
        {
            if (value >= 0)
                memory = value;
        }
    }

    // Конструктор смартфона
    public SmartPhone(string name, double price, string manufacturer, string category, int stockQuantity, double weight,
        int warrantyPeriod, int power, bool isWireless, string operatingSystem, int memory)
        : base(name, price, manufacturer, category, stockQuantity, weight, warrantyPeriod, power, isWireless)
    {
        OperatingSystem = operatingSystem;
        Memory = memory;
    }

    // Переопределение вывода информации о смартфоне
    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Гарантия: {WarrantyPeriod}, Мощность: {Power}, Беспроводной: {IsWireless}, ОС: {OperatingSystem}, Память: {Memory}");
    }

    // Зарядка смартфона
    public void Charge()
    {
        Console.WriteLine($"{Name} заряжается.");
    }

    // Подключение смартфона к интернету
    public void ConnectToInternet()
    {
        Console.WriteLine($"{Name} подключен к интернету.");
    }
}
```

```
// Простое наследование
Electronics laptop = new Electronics("Ноутбук", 80000, "Lenovo", "Электроника", 5, 2.4, 36, 120, false);
Clothing tshirt = new Clothing("Футболка", 1500, "Nike", "Одежда", 20, 0.3, "M", "Хлопок", "Лето");
Books book = new Books("Преступление и наказание", 700, "Эксмо", "Книги", 12, 0.5, "Ф. М. Достоевский", 672, "Роман");

// Сложное и множественное наследование
SmartPhone phone = new SmartPhone("Смартфон", 65000, "Samsung", "Электроника", 8, 0.2, 24, 25, true, "Android", 256);

// Вывод информации об электронике
Console.WriteLine("Информация об электронике:");
laptop.Display();
laptop.TurnOn();
laptop.Discount(10);
Console.WriteLine("После скидки:");
laptop.Display();

Console.WriteLine();

// Вывод информации об одежде
Console.WriteLine("Информация об одежде:");
tshirt.Display();
tshirt.TryOn();

Console.WriteLine();

// Вывод информации о книге
Console.WriteLine("Информация о книге:");
book.Display();
book.Read();

Console.WriteLine();

// Вывод информации о смартфоне
Console.WriteLine("Информация о смартфоне:");
phone.Display();
phone.TurnOn();
phone.Charge();
phone.ConnectToInternet();

Console.WriteLine();

// Взаимодействие объектов между собой через сравнение цены
Console.WriteLine("Взаимодействие объектов:");
laptop.ComparePrice(tshirt);
tshirt.ComparePrice(book);
phone.ComparePrice(laptop);

Console.WriteLine();

// Работа с наличием товара
Console.WriteLine("Проверка наличия товара:");
Console.WriteLine($"Ноутбук в наличии: {laptop.IsAvailable()}");
book.Restock(5);
Console.WriteLine($"Количество книг после пополнения: {book.StockQuantity}");
```
